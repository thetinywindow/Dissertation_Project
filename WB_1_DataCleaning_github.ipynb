---
The author of the analysis decided to validate answers based on the required time of responses. In detail, answers quicker than 3min and longer than 2hours are excluded.

---

```python
#kaggle initial data cleaning
####################################################

#transform reposnse time to numeric value
df_an['Time from Start to Finish (seconds)'] = pd.to_numeric(df_an['Time from Start to Finish (seconds)'])
df_an_num_res = len(df_an.index)

#spend less than 3min OR more than 2hours
df_an = df_an[df_an['Time from Start to Finish (seconds)']>180] #3min * 60sec = 180
df_an = df_an[df_an['Time from Start to Finish (seconds)']<7200] #2h * 60min * 60sec = 7200
df_an_num_res_new = df_an_num_res - len(df_an.index)

df_an['valid_response'] = 1

#outout
print('Number of excluded datapoints: ',df_an_num_res_new)
print('New data frame shape: ',df_an.shape)
```

---
Next, the job titles had to be standardtized for the following analysis; also important for later comparison between Kaggle and Glassdoor data.

-------

```python
#treat positions/ job levels
####################################################

df_an['Q5'] = df_an['Q5'].fillna('Other')

#drop students, others and non employed
df_an = df_an[df_an['Q5']!='Student']
df_an = df_an[df_an['Q5']!='Other']
df_an = df_an[df_an['Q5']!='Not employed']

#Business/Data Analyst
df_an.Q5[df_an['Q5']=='Business Analyst'] = 'Business/Data Analyst'
df_an.Q5[df_an['Q5']=='Data Analyst'] = 'Business/Data Analyst'

#Researcher
df_an.Q5[df_an['Q5']=='Research Scientist'] = 'Researcher'
df_an.Q5[df_an['Q5']=='Statistician'] = 'Researcher'

#Data/Software Engineer
df_an.Q5[df_an['Q5']=='Data Engineer'] = 'Data/Software Engineer'
df_an.Q5[df_an['Q5']=='Software Engineer'] = 'Data/Software Engineer'

print(df_an.groupby(df_an['Q5']).mean())
print(df_an.shape)
```

```python
#export csv for seperate visualization workbook
###################################

df_an.to_csv(r'/project/DATA/Kaggle_Survey/Data_Cleaning/CLEANED_valid_responses.csv', index=False)

###################################
```

---
Data cleaning commenced with the preperation for predictive modelling. Besides others this included target variable engineering, creating dummy variables, treating data anomalies and normalizing data. Please see the below code section for details.


---

```python
#copy previous data frame to new predictive dataframe
df_pred = df_an.copy()

#output
print(df_pred.shape)
df_pred.head(1)
```

```python
#target variabel engineering 
#############################

#rework target variable (Q10) by formatting string data
comp = df_pred.Q10.str.replace(',', '').str.replace('500000\+', '500000-500000').str.replace('$', '').str.split('-')
df_pred['yearly_comp_combined'] = comp

#split single string column in to bottom and top value column
marker = df_pred['yearly_comp_combined'].apply(pd.Series)
marker = marker.rename(columns = lambda x : 'mark_' + str(x))
marker.dropna(inplace=True)

#set types to numerical and form average of both bottom and top end data per instance
marker['mark_0'] = marker['mark_0'].astype(int)
marker['mark_1'] = marker['mark_1'].astype(int)
marker['result'] = (marker['mark_0']+marker['mark_1'])/2 #average of both marker 1 and 2

#include compensation in target variable format
df_pred['yearly_comp_numerical'] = marker['result'] #set yearly compensation to previously engineered values
df_pred['yearly_comp_numerical'] = df_pred['yearly_comp_numerical'].fillna(0)
df_pred = df_pred.drop('yearly_comp_combined', axis = 1)

#drop rows with compensation = 0
#print(df_pred.shape)
#df_pred = df_pred[df_pred['yearly_comp_numerical']!=0]

#output
print(df_pred.shape)
df_pred[['Q10', 'yearly_comp_numerical']].head(3)
```

```python
#data compensation validation (removing responses with error in compensation data field)
#########################

indexNames_BA = df_pred.loc[((df_pred['Q5']=='Data Scientist')&(df_pred['yearly_comp_numerical']<=2500)),:].index
df_pred.drop(indexNames_BA , inplace=True)

indexNames_DS = df_pred.loc[((df_pred['Q5']=='Data Scientist')&(df_pred['yearly_comp_numerical']<=2500)),:].index
df_pred.drop(indexNames_DS , inplace=True)

indexNames_SE = df_pred.loc[((df_pred['Q5']=='Data/Software Engineer')&(df_pred['yearly_comp_numerical']<=2500)),:].index
df_pred.drop(indexNames_SE , inplace=True)

indexNames_PM = df_pred.loc[((df_pred['Q5']=='Data/Software Engineer')&(df_pred['yearly_comp_numerical']<=2500)),:].index
df_pred.drop(indexNames_PM , inplace=True)

indexNames_RS = df_pred.loc[((df_pred['Q5']=='Data/Software Engineer')&(df_pred['yearly_comp_numerical']<=2500)),:].index
df_pred.drop(indexNames_RS , inplace=True)

#output
print(df_pred.shape)
```
